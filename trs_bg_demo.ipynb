Compute TRS for image pair 

Case 1: single translation

```python
op_res = (256, 256)
```

```python
img_rgb = resize(imread('data/synth_trs_trans.png'), op_res, mode = 'reflect')
img1 = img_rgb[:,:,0]
img2 = img_rgb[:,:,2]
tmap, rsmap = trs_pair.trs_pair(img1, img2)
plt.figure(figsize = (15,45))
plt.subplot(131)
plt.imshow(img_rgb)
plt.subplot(132)
plt.title('T-map')
plt.imshow(tmap)
plt.subplot(133)
plt.title('RS-map')
plt.imshow(rsmap)
plt.show()
```

background subtraction

```python
for i in range(9):
    temp = resize(np.mean(imread('data/runningman_river/' + '%0*d' % (2, i+1) + '.png', as_gray = True),axis = 2), op_res, mode = 'reflect')/255.0
    if i == 0:
        vid = temp
    else:
        vid = np.dstack((vid, temp))
plt.figure(figsize = (15,45))
plt.subplot(131)
plt.imshow(vid[:,:,0])
plt.subplot(132)
plt.imshow(vid[:,:,1])
plt.subplot(133)
plt.imshow(vid[:,:,2])
plt.show()

vid_res = np.shape(vid)
```

```python
bg = background(vid_res)

kvar = 5**2*1e-4
vid_fg = bg.med_filter(vid, kvar)

plt.figure(figsize = (15,45))
plt.subplot(131)
plt.imshow(vid_fg[:,:,0])
plt.subplot(132)
plt.imshow(vid_fg[:,:,1])
plt.subplot(133)
plt.imshow(vid_fg[:,:,2])
plt.show()

```
