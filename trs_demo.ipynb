```python
from utils.trs import lp_precomp
from utils.trs import trs
import numpy as np
from skimage.io import imread
from skimage.transform import resize
import matplotlib.pyplot as plt
```

Compute TRS for image pair 

Case 1: single translation

```python
op_res = (256, 256)
```

```python
img_rgb = resize(imread('data/synth_trs_trans.png'), op_res, mode = 'reflect')
img1 = img_rgb[:,:,0]
img2 = img_rgb[:,:,2]
tmap, rsmap = trs_pair.trs_pair(img1, img2)
plt.figure(figsize = (15,45))
plt.subplot(131)
plt.imshow(img_rgb)
plt.subplot(132)
plt.title('T-map')
plt.imshow(tmap)
plt.subplot(133)
plt.title('RS-map')
plt.imshow(rsmap)
plt.show()
```
